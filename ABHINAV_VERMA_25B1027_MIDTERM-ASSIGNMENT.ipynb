# üìä Data Visualization

## üìå Plot 1: Histogram of Study Hours

```python
plt.figure(figsize=(6,4))
plt.hist(df['Study_Hours'], bins=10)
plt.xlabel('Study Hours')
plt.ylabel('Frequency')
plt.title('Distribution of Study Hours')
plt.show()
```

### üìù Observation:
Most students study for an average number of hours. Only a few students study very less or very high hours. This shows that study habits are mostly similar among students.

## üìå Plot 2: Bar Chart of Gender

```python
plt.figure(figsize=(6,4))
df['Gender'].value_counts().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.title('Gender Distribution')
plt.show()
```

### üìù Observation:
The number of male and female students is slightly different. However, the overall distribution looks fairly balanced.

## üìå Plot 3: Scatter Plot (Study Hours vs Math Score)

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Study_Hours'], df['Math_Score'])
plt.xlabel('Study Hours')
plt.ylabel('Math Score')
plt.title('Study Hours vs Math Score')
plt.show()
```

### üìù Observation:
Students who study more hours generally score higher in math. Still, the points are not perfectly aligned, which means other factors may also affect performance.
